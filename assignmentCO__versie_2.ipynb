```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
CO2_data = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
FD = CO2_data[1]
FD.columns = ["Country", "MTCO_1990", "MTCO_2005", "MTCO_2017", "2017_world_%", "2017_perc_change","Per_land_area", "Per_capita", "incl_LUCF", "Excl_LUCF"]
FD = FD[ FD["Country"].str.contains ("World|European Union")== False]

FD
```

```python
Large_countries = FD.nlargest(5,["MTCO_1990","MTCO_2005","MTCO_2017"])
Large_countries
```
